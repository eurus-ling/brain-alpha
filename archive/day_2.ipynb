```python
import pandas as pd
import requests

def get_datafields(s, searchScope, dataset_id: str = '', search: str = ''):
    instrument_type = searchScope['instrumentType']
    region = searchScope['region']
    delay = searchScope['delay']
    universe = searchScope['universe']
    
    if len(search) == 0:
        url_template = "https://api.worldquantbrain.com/data-fields?" \
            f"&instrumentType={instrument_type}" \
            f"&region={region}&delay={str(delay)}&universe={universe}&dataset.id={dataset_id}&limit=50" \
            "&offset={x}"
        count = s.get(url_template.format(x=0)).json()['count']
    else:
        url_template = "https://api.worldquantbrain.com/data-fields?" \
            f"&instrumentType={instrument_type}" \
            f"&region={region}&delay={str(delay)}&universe={universe}&limit=50" \
            f"&search={search}" \
            "&offset={x}"
        count = 100

    datafields_list = []
    for x in range(0, count, 50):
        datafields = s.get(url_template.format(x=x))
        datafields_list.append(datafields.json()['results'])
    
    datafields_list_flat = [item for sublist in datafields_list for item in sublist]
    
    datafields_df = pd.DataFrame(datafields_list_flat)
    return datafields_df
```

```python
searchScope = {'region': 'USA', 'delay': 1, 'universe': 'TOP3000','instrumentType': 'EQUITY'}
fundamental6 = get_datafields(s = sess, searchScope = searchScope, dataset_id = 'fundamental6')
```

```python
fundamental6 = fundamental6[fundamental6['type'] == 'MATRIX']
fundamental6.head()
```

```python
datafields_list_fundamental6 = fundamental6['id'].values
```

```python
alpha_list = []

for datafield1 in datafields_list_fundamental6:
    for datafield2 in datafields_list_fundamental6:
        alpha_expression = f"group_rank({datafield1}/{datafield2}, subindustry)"
        simulation_data = {
            'type': 'REGULAR',
            'settings': {
                'instrumentType': 'EQUITY',
                'region': 'USA',
                'universe': 'TOP3000',
                'delay': 1,
                'decay': 0,
                'neutralization': 'SUBINDUSTRY',
                'truncation': 0.08,
                'pasteurization': 'ON',
                'unitHandling': 'VERIFY',
                'nanHandling': 'ON',
                'language': 'FASTEXPR',
                'visualization': False,
            },
            'regular': alpha_expression
        }
        alpha_list.append(simulation_data)

print(f"there are {len(alpha_list)} Alphas to simulate")
```

```python
from time import sleep

for alpha in alpha_list:
    sim_resp = sess.post(
        'https://api.worldquantbrain.com/simulations',
        json=alpha,
    )
    
    try:
        sim_progress_url = sim_resp.headers['location']
        while True:
            sim_progress_resp = sess.get(sim_progress_url)
            retry_after_sec = float(sim_progress_resp.headers.get("Retry-After", 0))
            if retry_after_sec == 0:  # simulation done!
                break
            sleep(retry_after_sec)
        
        alpha_id = sim_progress_resp.json()["alpha"]  # the final simulation result
        print(alpha_id)
        
    except:
        print("no location, sleep for 10 seconds and try next alpha")
        sleep(10)
```

```python
import time
import requests
from datetime import datetime, timedelta

def format_seconds(sec):
    return str(timedelta(seconds=int(sec)))

def run_simulation(sess, alpha, idx, total, times, min_check_interval=40, max_retries=3):
    start_time = time.time()
    print(f"\n[{datetime.now().strftime('%Y-%m-%d %H:%M:%S')}] "
          f"开始处理 Alpha {idx}/{total} (进度 {idx/total*100:.2f}%)")

    for attempt in range(max_retries):
        try:
            sim_resp = sess.post(
                "https://api.worldquantbrain.com/simulations",
                json=alpha,
                timeout=30
            )

            if sim_resp.status_code != 201 or "location" not in sim_resp.headers:
                print(f"[WARN] 提交失败，状态码 {sim_resp.status_code}，重试 {attempt+1}/{max_retries}")
                time.sleep(10)
                continue

            sim_progress_url = sim_resp.headers["location"]

            # 轮询
            while True:
                sim_progress_resp = sess.get(sim_progress_url, timeout=30)
                retry_after_sec = float(sim_progress_resp.headers.get("Retry-After", 0))

                if retry_after_sec == 0:  # done
                    data = sim_progress_resp.json()
                    alpha_id = data.get("alpha")
                    duration = time.time() - start_time
                    if alpha_id:
                        times.append(duration)
                        avg_time = sum(times) / len(times)
                        remaining = (total - idx) * avg_time
                        print(f"[{datetime.now().strftime('%Y-%m-%d %H:%M:%S')}] "
                              f"完成 Alpha {idx}/{total} ✅ ID={alpha_id} "
                              f"(耗时 {format_seconds(duration)}, "
                              f"预计剩余 {format_seconds(remaining)})")
                        return alpha_id
                    else:
                        print("[ERROR] 仿真完成但未返回 alpha_id")
                        return None

                # 控制最小检查间隔（减少刷屏）
                wait_sec = max(retry_after_sec, min_check_interval)
                print(f"[INFO] Alpha {idx}/{total} 正在运行中... "
                      f"下次检查 {wait_sec:.0f} 秒后")
                time.sleep(wait_sec)

        except requests.exceptions.RequestException as e:
            print(f"[ERROR] 网络请求失败: {e}，重试 {attempt+1}/{max_retries}")
            time.sleep(10)

    print(f"[FATAL] Alpha {idx}/{total} 重试超过上限，放弃。")
    return None


# 批量运行
total = len(alpha_list)
all_start = time.time()
times = []

print(f"[{datetime.now().strftime('%Y-%m-%d %H:%M:%S')}] 开始批量运行 {total} 个 Alphas")

for idx, alpha in enumerate(alpha_list, 1):
    alpha_id = run_simulation(sess, alpha, idx, total, times, min_check_interval=40)
    if not alpha_id:
        print(f"[FAIL] Alpha {idx}/{total} 无法获取结果，跳过。")

all_end = time.time()
print(f"\n[{datetime.now().strftime('%Y-%m-%d %H:%M:%S')}] "
      f"全部完成 ✅ 总耗时 {format_seconds(all_end - all_start)}")
```
